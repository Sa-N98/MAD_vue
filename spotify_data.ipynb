```python
def artist(name):
    artist_name = name

    # Spotify Search API endpoint
    search_url = 'https://api.spotify.com/v1/search'

    # Parameters for artist search
    search_params = {
        'q': artist_name,
        'type': 'artist',
        'limit': 1  # Limiting to 1 result for simplicity
    }

    # Headers with the access token
    search_headers = {
        "Authorization": f"Bearer {access_token}"
    }

    # Make the artist search request
    search_response = requests.get(search_url, params=search_params, headers=search_headers)

    # Check if the search request was successful
    if search_response.status_code == 200:
        result=search_response.json()
        return result
```

```python
def album(id):
    search_headers = {
                            "Authorization": f"Bearer {access_token}"
                        }
    albums_url = f"https://api.spotify.com/v1/artists/{id}/albums"
    albums_params = {
        'include_groups': 'album,single',  # Include albums and singles
        'limit': 50  # Limiting to 50 albums for simplicity (adjust as needed)
    }

        # Make the albums request
    albums_response = requests.get(albums_url, params=albums_params, headers=search_headers)
    if albums_response.status_code == 200:
        result=albums_response.json()
        return result

```

```python
def tracks(id):
    search_headers = {
                            "Authorization": f"Bearer {access_token}"
                        }
    tracks_url = f"https://api.spotify.com/v1/albums/{id}/tracks"
    tracks_response = requests.get(tracks_url, headers=search_headers)

    if tracks_response.status_code == 200:
        result=tracks_response.json()
        return result
```
